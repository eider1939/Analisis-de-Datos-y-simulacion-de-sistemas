Planteamiento:

En el juego de CRAPS un jugador lanza un par de dados una o más veces hasta que gana o pierde. 

Hay dos maneras de ganar este juego:

Obtener 7 u 11 en el primer lanzamiento.

Obtener 4, 5, 6, 8, 9 ó 10 en el primer lanzamiento y repetir lo obtenido en el primer lanzamiento antes de obtener un 7.

Hay dos formas de perder:

Obtener 2, 3 ó 12 en el primer lanzamiento.

Obtener 7 en un lanzamiento posterior, antes de obtener lo lanzado en la primera ocasión.




Suponga que juega CRAPS 3 veces. Cada intento, debe pagar $1 para iniciar y recibe $2 si gana.

 

```python
import random 
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats

def dado():
    r=random.random()
    if r <= 1/6:
        return 1
    elif r<= 2/6:
        return 2
    elif r<=3/6:
        return 3
    elif r<=4/6:
        return 4
    elif r<=5/6:
        return 5
    else:
        return 6

def craps():
  utilidad = -1  # la utilidad es -1 porque pagó 1 por jugar
  l = 1
  Lanzamiento1=dado()+dado() # L1 es el resultado del primer lanzamiento
  #print("lanzamineto1 ",Lanzamiento1)
  # Pierde si L1 es 2, 3 o 12
  L1_pierde=[2,3,12]
  L1_gana=[7,11]
  if Lanzamiento1 in L1_pierde:  
      return utilidad
  if Lanzamiento1 in L1_gana:  # gana si L1 es 7 u 11
      utilidad = utilidad + 2
      return utilidad
  # continúa para L1=4, 5, 6, 8, 9, 10
  l = l + 1 
  Lanzamiento2 = dado()+dado()
  #print("lanzamineto2 ",Lanzamiento2)
  # mientras L2 no sea 7 se puede continuar.
  while Lanzamiento2 != 7:
    # Gana si L2 es igual a L1
      if Lanzamiento2 == Lanzamiento1:
          utilidad = utilidad + 2
          break
      else:
          l=l+1
          Lanzamiento2 = dado()+dado()
          #print("lanzamineto2 nuevo",Lanzamiento2 )
            
  return utilidad

#craps()
```

```python
#ingresar el número N de veces (un entero) que se jugará craps()
N=int(input('¿Cuántas veces va a jugar CRAPS? '))
iteraciones=0
g = []  
while iteraciones < N:
  #print("juego nuevo ", iteraciones)
  num1=craps()
  g += [num1]
  iteraciones+=1

gano=0
for i in g:
  if i==1:
    gano+=1
```

```python
print("De ",N, " Juegos gano ",gano)
p=gano/N
utilidad_esperada=p*2+(1-p)*0-1
print("Probabilidad de ganar es ", gano/N)
print("Utilidad esperada: ", utilidad_esperada)

valor_cobrar=p*2-utilidad_esperada
print("El valor justo que se deberia cobrar es: ", valor_cobrar)
```

```python
# cálculo de la media y la desviación estándar de las ganancias
gprom=np.mean(g)
gdesv=np.std(g)
# los grados de libertad se calculan de acuerdo con la fórmula como:
df=N-1
# El estadístico t crítico es:
tc= scipy.stats.t.ppf(0.975,df)
print(tc)
# la mitad del intervalo es
mic=tc*gdesv/(N**0.5)
print("media=",gprom, "+-",mic, "95% nivel confianza")
# este método usa la función intervalo de confianza
sem=scipy.stats.sem(g)
ic=scipy.stats.t.interval(0.95,df,gprom,sem)
print("media=",gprom,"Intervalo confianza 95%=",ic)
```
