A los vendedores de una tienda de bicicletas les pagan un bono si venden más de 4 bicicletas diarias. La probabilidad de vender más de 4 bicicletas es sólo 0.4. Si el número de bicicletas vendido es mayor que 4, la distribución de las ventas es la siguiente:

La tienda tiene 4 modelos de bicicletas y el bono depende del modelo vendido. Para el modelo A el bono es de $10; 40% de las bicicletas vendidas son de ese tipo. El modelo B representa el 35% de las ventas y paga un bono de $15. El modelo C paga un bono de $20 y representa el 20% de las ventas y el modelo D tiene un bono de $25 y representa sólo el 5% de las ventas. 



Actividad:

Crear un programa en Python o Excel que calcule el valor esperado del bono que un vendedor puede obtener en un día.


```python
import random 
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem, t
from scipy import mean
```

```python

def prob_vendidas():
  r = random.random()
  if r <= 0.6:
    return 0
  elif r <= 0.74:
    return 5
  elif r <= 0.92:
    return 6
  elif r <= 0.98:
    return 7
  else:
    return 8

def num_bic():
  r1 = random.random()
  if r1 <= 0.4:
    return 10
  elif r1 <= 0.75:
    return 15
  elif r1 <= 0.95:
    return 20
  else:
    return 25


def total_bonos_dia():
  total_prob_vendidas = prob_vendidas()
  totalBonos = 0
  for i in range(0, total_prob_vendidas):
    totalBonos += num_bic()
  return totalBonos

```

Con base en 100 simulaciones, para una precisión de D=1, calcule el número necesario de iteraciones.


```python
lista_dia_bonos_100=[]
N=100
for i in range(0,N):
  lista_dia_bonos_100+=[total_bonos_dia()]
print(str(lista_dia_bonos_100))
w=np.std(lista_dia_bonos_100)
n=(1.96*w)**2
print("para una presicion de D=1 se necesitas ",n," iteraciones")
```

Con base en al menos 10000 iteraciones,
Construir un histograma del valor del bono y analizarlo. 

```python
lista_dia_bonos=[]
N=10000
for i in range(0,N):
  lista_dia_bonos+=[total_bonos_dia()]

print(str(lista_dia_bonos))
```

```python
a = np.hstack(lista_dia_bonos)
plt.hist(a, bins='auto')
plt.title("Distribucion de bonificaciones en: 10000 dias")
plt.show()
```

Calcular estadísticas descriptivas de las utilidades diarias: mínimo, máximo, media, mediana y desviación estándar e interpretarlas. 


```python
summary = pd.DataFrame(lista_dia_bonos)
summary.describe()
```

```python
media = summary.mean()
std = summary.std()
print("Media: ", (round(media,2)))
print("Desviación estándar: ", (round(std,2)))
```

Calcular un intervalo de confianza para el valor esperado del bono, considerando un valor de significancia de 5%. 


```python
confidence = 0.95
n = len(summary)
std_err = sem(summary)
h = std_err * t.ppf((1 + confidence) / 2, n - 1)      # Calculamos el intervalo de confianza

lim_inf = media - h
lim_sup = media + h
print("h = " , (h))
print ("Int. de confianza para los bonos diarios: ", (round(lim_inf,2)), " -- ",(round(lim_sup,2)))
```
