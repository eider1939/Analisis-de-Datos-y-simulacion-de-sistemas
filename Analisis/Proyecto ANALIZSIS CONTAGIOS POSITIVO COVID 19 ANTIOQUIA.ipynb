```python
#importamos las librerias que necesitamos
import pandas as pd
import datetime
import numpy as np
import matplotlib as plt
import matplotlib
from datetime import datetime
import matplotlib.pyplot as plt
color = ['blue','green','purple', 'brown','black','tomato',
         'red', 'gray','pink']
```

```python
#cargamos el archivo csc este archivo debe subirse a la carpeta de archivos del colabority o estar en la misma carpeta del archivo .ipynb
Datos_Covid = "Casos_positivos_de_Covid-19_en_el_departamento_de_Antioquia.csv"
DatosCovid= pd.read_csv(Datos_Covid) #lee el archivo csv  y se crea un dataframe crea un tabla
DatosCovid
```

```python
DatosCovid.columns #vemos las columnas que tiene el archivo csv
```

SOLUCION:

1) un analisis por año y mes

como vimos las columnas tiene un dato tipo FECHA el cual toca manipular para poder realizar el analisis por MES Y AÑO

para este se agragaran tres columnas FECHA AÑO MES

```python
#agragamos 3 columnas al archivo csv "Fecha" sale de la columna "fecha reporte web"
DatosCovid['Fecha'] = pd.to_datetime(DatosCovid['fecha reporte web'], errors='coerce')
DatosCovid['Año'] = DatosCovid['Fecha'].dt.year  #se agraga la columna "Año"
DatosCovid['Mes'] = DatosCovid['Fecha'].dt.strftime('%m') #se agrega la columna "Mes"
DatosCovid.columns  #comprovamos que si se esten las nuevas columnas
```

```python
#extraemos la cantidad de contagiados por año CONTANDO SUS VALORES DE FORMA ASENDENTE
cant_contagiados_año=DatosCovid['Año'].value_counts().sort_index(ascending=True)
cant_contagiados_año
```

```python
cant_contagiados_año.describe()
```

```python
#grafica de casos positivos por Año en el Departamento de Antioquia
plt.plot(cant_contagiados_año.index[:12], cant_contagiados_año.values[:12], marker= '*', markersize=8.0, linewidth=3.0, 
         linestyle='solid', color='red')
plt.xlabel('AÑO', fontsize='15')
plt.ylabel('Numero de Contagios', fontsize='10')
plt.title("Contagios por AÑO", fontsize='12')
plt.grid(color='grey', linestyle='solid', linewidth=1.5, alpha=0.5)
```

En la grafica podemos ver como el numero de contagios crecio lineal mente en el tiempo hasta mediados del 2021 luego decrecio,

con un media de 305478
un minimo de 122985.000000 y un maximo 531845.000000

```python
#extraemos la cantidad de contagiados por mes CONTANDO SUS VALORES DE FORMA ASENDENTE
cant_contagiados_mes=DatosCovid['Mes'].value_counts().sort_index(ascending=True)
cant_contagiados_mes
```

```python
cant_contagiados_mes.describe()
```

```python
#grafica de casos positivos por mes en el Departamento de Antioquia
plt.plot(cant_contagiados_mes.index[:12], cant_contagiados_mes.values[:12], marker= '*', markersize=8.0, linewidth=3.0, 
         linestyle='solid', color='red')
plt.xlabel('Mes', fontsize='15')
plt.ylabel('Numero de Contagios', fontsize='10')
plt.title("Contagios por Mes", fontsize='12')
plt.grid(color='grey', linestyle='solid', linewidth=1.5, alpha=0.5)
```

como vamos la grafica tubo varios picos en los cuales los contagios se dispararon,teniendo como mes mas algido al mes 01 "ENERO" esto se pude justificar ya que en las epocas decenbrinas las perosnas tenien menos precuacion y acataban menos als restriciones

2) realizar un alalisis por edad


```python
contagios_edad=DatosCovid['Edad'].value_counts().sort_index(ascending=True)
contagios_edad
```

3) analisis por sexo de la poblacion por mes.

```python
F = DatosCovid[(DatosCovid['Sexo'] == 'F')]       
cant_contagiados_F=F['Mes'].value_counts().sort_index(ascending=True)  
M = DatosCovid[(DatosCovid['Sexo'] == 'M')]       
cant_contagiados_M=M['Mes'].value_counts().sort_index(ascending=True) 
```

```python
cant_contagiados_F=F['Edad'].value_counts().sort_index(ascending=True)       
cant_contagiados_M=M['Edad'].value_counts().sort_index(ascending=True)  
plt.figure(figsize=(10,6))
plt.bar(cant_contagiados_F.index[:112], cant_contagiados_F.values[:112], color = "purple",label='F')
plt.bar(cant_contagiados_M.index[:112], cant_contagiados_M.values[:112], color = 'blue',label='M')
plt.legend()
plt.xlabel("EDAD")
plt.ylabel("Cantidad de contagios")
plt.title("Contagios por Edad y Sexo") 
plt.show()
```

4) analisis respecto a la cantidad de contagio que a se vivieron enfocado  especificamente en el valle de aburra que es el lugar donde habito.

```python
cant_mes_municipio = DatosCovid['Nombre municipio'].unique()       #extraemos el numbre de los municipios
lista_municipios_aburra=['MEDELLIN', 'ITAGUI', 'BELLO','LA ESTRELLA','MEDELLiN', 
                         'Medellin','SABANETA','COPACABANA','GIRARDOTA']
lista_mun=[]   
describe_mun=[]                                                       # se crea un amtrix lista_mun
for i in lista_municipios_aburra:                                          #ciclo para recorrer la lista con el nombre de los municipios
  l=[]
  Mun = DatosCovid[(DatosCovid['Nombre municipio'] == i)]       #esrraemos de la tabla las filas que tenga el nombre del municipio
  canti_contagiados_mes=Mun['Mes'].value_counts().sort_index(ascending=True)  #contamos los coantagiados por mes
  s=canti_contagiados_mes.describe()
  describe_mun.append(s)
  l.append(i)                                                         # se agrega a la lista l el nombre del municipio
  l.append(canti_contagiados_mes)                                     #se agrega el arreglos de la cantidad de contagiados por mes a l
  lista_mun.append(l)   
```

```python
#se recorre la matrix lista_mun se estre el numbre y los datos de contagios por mes
#para realizar las respectivas graficas para cada municipio
for i in range(0,len(lista_mun)):   #ciclo que recorre
  print(  )
  print("Descripcion estadistica Municipio ",str(lista_mun[i][0]))
  print( )
  print(describe_mun[i])
  plt.bar(lista_mun[i][1].index[:12], lista_mun[i][1].values[:12], color = color[:12])
  plt.xlabel("Mes")
  plt.ylabel("Cantidad de contagios")
  plt.title("Contagios por mes en "+ lista_mun[i][0])  #re va dando el respectivo nombre de cada municipio a cada grafica
  plt.show()
```
