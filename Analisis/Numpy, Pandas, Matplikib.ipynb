PROBLEMA 2 (25%). Pandas.

Puede reutilizar código del problema 2 del parcial#2. Su programa genera la lista de primos entre 10 y
1_000_000 y luego identifica el último digito de cada primo y genera un diccionario con las claves “ uno”,
“tres”, “siete”, “nueve”. El valor de cada key o clave es una lista con los primos que finalizan en el dígito que
indica su clave. EJM. “uno”:[11,31,41 etc] “nueve”:[19,29….etc]. Investigue como generar un dataframe a
partir de este diccionario. Luego use la capacidad de la librería pandas para graficar y realice una gráfica de
barras o barplot, eje x es la clave y el eje y es el número de elementos de la lista que le corresponde. Son
similares las alturas entre las barras? Qué se puede concluir?


```python
lista_primos=[]
for i in range(10,100000):
  verificador=True
  for n in range(2,i):
    if i % n == 0:
      verificador=False
  if verificador==True:
    lista_primos.append(i)
```

```python
uno=[]
tres=[]
siete=[]
nueve=[]
for j in lista_primos:
  b=str(j)
  a=int(b[len(b)-1])
  if a==1:
    uno.append(j)
  elif a==3:
    tres.append(j)
  elif a==7:
    siete.append(j)
  elif a==9:
    nueve.append(j)
print(len(uno)," ",len(tres)," ",len(siete)," ",len(nueve))
```

```python
ll=[len(uno),len(tres),len(siete),len(nueve)]
D=max(len(uno),len(tres),len(siete),len(nueve))
def l_ceros(L,D):
  for i in range(len(L),D):
    L.append(0)
  return L

uno=l_ceros(uno,D)
tres=l_ceros(tres,D)
siete=l_ceros(siete,D)
nueve==l_ceros(nueve,D)
```

```python
dicc_primos={'Uno':uno,'Tres':tres,"Siete":siete,"Nueve":nueve}
primos=pd.DataFrame(dicc_primos)
primos
```

```python
X=primos.columns.values
fig,ax=plt.subplots()
ax.bar(X,ll)
#plt.yticks(range(150,320))
## Creamos Gráfica
plt.show()
```

PROBLEMA 3. (Valor 25%) Python Puro

Sorprendentemente solo hay tres números iguales a la suma de la cuarta potencia de sus dígitos:
1634 = 14 + 64 + 34 + 44 , 8208 = 84 + 24 + 04 + 84 , 9474 = 94 + 44 + 74 + 44 .
Note que 1 = 14 al ser un único dígito, no es una suma propiamente y no se incluye. Por otro lado, la suma
de estos tres particulares números es 1634 + 8208 + 9474 = 19316.
Escribir un programa que hallará la suma de todos los números que son iguales a la suma de las potencias
quintas de sus dígitos. Existe un modo teórico/analitico para determinar el número máximo hasta el cual buscar?

```python
def quinta_potencia(X):
  s=str(X)                     #convierte el numero en string para poder recorrelos como cadena
  suma=0
  for i in s:                 #se itera en s
    d=int(i)**5               #se halla la potencia 5 del digito
    suma+=d                   #se suma a las potencias de 5 ya halladas
  if suma==X:                 #verifica si son iguales
    return True                
```

```python
suma_5=0
for i in range(2,295245):                #itera en un rago definido
  s=quinta_potencia(i)                   #llama a la funcion 
  if s:
    suma_5+=i                           #suma los numeros que cumplen la condicion
print("la suma de todos los números que son iguales \na la suma de las potencias quintas de sus dígitos es: ",suma_5)
```

C) Para terminar grafique en un panel horizontal 1x2 una a la lado de la otra, los datos “x vs y”

Solucion A

```python
from numpy import random
x = np.linspace(0,10, 30)
a=random.randint(0,2)
b=np.linspace(-2,2, 30)
F=(x**2)+a*x+b
plt.plot(F)
plt.show()
```

```python
z11 = np.polyfit(x, F, 2)
p11 = np.poly1d(z11)
print("Polinomio grado 2 ")
print(p11)
print( )

yvals = p11(x) 
www=plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plot1 = plt.plot(x, F, '*', label='original values')
plot2 = plt.plot(x, yvals, 'r', label='polyfit values')

plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.legend(loc=4)  # Especifique la ubicación de la leyenda, el lector puede ayudar a su uso
plt.title('polyfitting Grade 2')

plt.subplot(1, 2, 2)
z22 = np.polyfit(x, F, 3)
p1 = np.poly1d(z22)

# Imprime el polinomio ajustado en la pantalla
print("Polinomio grado 3 ")
print(p1)


yvals = p1(x)  # También puedes usar yvals = np.polyval (z1, x)

plot1 = plt.plot(x, F, '*', label='original values')
plot2 = plt.plot(x, yvals, 'r', label='polyfit values')

plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.legend(loc=4)  # Especifique la ubicación de la leyenda, el lector puede ayudar a su uso
plt.title('polyfitting Grade 3')
plt.show()
www.savefig('Graficas punto 4A.png')  #guarlas las graficas del punto 4A
pdf = PdfPages('Graficas punto 4A.pdf')
pdf.savefig(www)
pdf.close()
```
